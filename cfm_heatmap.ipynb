## Create sklearn ConfusionMatrixDisplay plot for IAA
imports true and annotated values from ud_iaa.py

saves plot to outfile

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
from ud_iaa import es_y_ud_anno, es_y_my_anno, pt_y_ud_anno, pt_y_my_anno
```

```python
print(set(es_y_my_anno))
```

```python
print(set(pt_y_my_anno))
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ConfusionMatrixDisplay.from_predictions(es_y_ud_anno, es_y_my_anno, ax=ax, cmap='RdPu')
plt.savefig('ancora_cfm.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ConfusionMatrixDisplay.from_predictions(pt_y_ud_anno, pt_y_my_anno, ax=ax, cmap='RdPu')
plt.savefig('bosque_cfm.png')
plt.show()
```
